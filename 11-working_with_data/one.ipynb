```python
import sqlite3
```

```python
connection = sqlite3.connect('exaple.db')
connection
```

```python
cursor = connection.cursor()
```

```python
cursor.execute("""
create table if not exists employees(
	id Integer Primary Key,
	name Text Not Null,
	age Integer,
	department text)
""")

## commit the changes
connection.commit()
```

```python
cursor.execute("""
select * from employees
""")
```

```python
## insert data in sqlite3

cursor.execute(
""" 
insert into employees(name, age, department)
values ('ben', 21, 'data engineer')
"""
)


cursor.execute(
""" 
insert into employees(name, age, department)
values ('tom', 20, 'iem')
"""
)


cursor.execute(
""" 
insert into employees(name, age, department)
values ('pond', 22, 'business man')
"""
)

connection.commit()
```

```python
# querying
cursor.execute(
	''' 
select * from employees
'''
)
rows=cursor.fetchall()

for row in rows:
	print(row)
```

```python
# update data
cursor.execute(
	''' 
update employees
set age = 34
where name = 'tom' '''
)
connection.commit()
```

```python
cursor.execute(
	''' 
select * from employees
'''
)
rows=cursor.fetchall()

for row in rows:
	print(row)
```

```python
# deleting data from talbe

cursor.execute(
	''' 
delete from employees
	where name='tom'
'''
)
connection.commit()
```

```python
cursor.execute(
	''' 
select * from employees
'''
)
rows=cursor.fetchall()

for row in rows:
	print(row)
```

```python
# working with sales data

connection = sqlite3.connect('sales_data.db')

cursor.execute(
	'''
	create table if not exists sales(
	id integer primary key,
	date text not null,
	product text not null,
	sales integer,
	region text)
'''
)
connection.commit()
```

```python
sales_data = [
    ('2023-01-01', 'Product1', 100, 'North'),
    ('2023-01-02', 'Product2', 200, 'South'),
    ('2023-01-03', 'Product1', 150, 'East'),
    ('2023-01-04', 'Product3', 250, 'West'),
    ('2023-01-05', 'Product2', 300, 'North')
]
cursor.executemany(
	'''
insert into sales(date, product, sales, region)
values(?, ?, ?, ?)
''', sales_data
)
connection.commit()
```

```python
# close the connection
connection.close()
```
