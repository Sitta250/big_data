```python
import sqlite3
```

```python
connection = sqlite3.connect('exaple.db')
connection
```

```python
cursor = connection.cursor()
```

```python
cursor.execute("""
create table if not exists employees(
	id Integer Primary Key,
	name Text Not Null,
	age Integer,
	department text)
""")

## commit the changes
connection.commit()
```

```python
cursor.execute("""
select * from employees
""")
```

```python
## insert data in sqlite3

cursor.execute(
""" 
insert into employees(name, age, department)
values ('ben', 21, 'data engineer')
"""
)


cursor.execute(
""" 
insert into employees(name, age, department)
values ('tom', 20, 'iem')
"""
)


cursor.execute(
""" 
insert into employees(name, age, department)
values ('pond', 22, 'business man')
"""
)

connection.commit()
```
